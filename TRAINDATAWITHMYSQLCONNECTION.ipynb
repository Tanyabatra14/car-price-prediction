# WHAT IS UNIVARIATE ANALYSIS

# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
df = df.replace({"Survived":{0:"Dead",1:"Alive"}})
df['Survived'].replace({0:"Dead",1:"Alive"}).value_counts().plot(kind='bar',rot=0)
```

# HOW MANY PASSENGERS WERE IN FIRST CLASS,SECOND CLASS AND THIRD CLASS PLOT THOSE FIGURES ON GRAPH

```python
df["Pclass"].replace({1:"First",2:"Second",3:"Third"}).value_counts().plot(kind='bar',rot=0,xlabel="Class")
```

# NUMBER  OF MALE AND FEMALE PASSANGERS

```python
df["Sex"].value_counts().plot(kind='bar',rot=0)
```

# BIVARIANT ANALSIS

# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
df.groupby(["Survived","Sex"]).size().unstack().plot(kind='bar',stacked=True)
```

# WHICH PASSENGER CLASS HAS BETTER CHANCE OF SURVIVAL(FIRST,SECOND,OR THIRD CLASS)?

```python
df.groupby(["Survived","Pclass"]).size().unstack().plot(kind='bar',stacked=True)
```
